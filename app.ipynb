### Soal 1: Statistik Dasar Dataset

```python
def mean(data: list[float]) -> float:
  if len(data) == 0:
    raise ValueError("List tidak boleh kosong")
  total = 0
  for i in range(len(data)):
    total = total + data[i]
  return float(total/len(data))

def median(data: list[float]) -> float:
  med = 0
  if len(data) == 0:
    raise ValueError("List tidak boleh kosong")
  
  sorted_data = sorted(data)
  n = len(sorted_data)
  mid = n // 2 
  
  if n % 2 == 0:
    left_mid = mid - 1
    med = (sorted_data[left_mid] + sorted_data[mid]) / 2.0
  else:
    med = sorted_data[mid]
        
  return float(med)

def mode(data: list[float]) -> float:
  if len(data) == 0:
    raise ValueError("List tidak boleh kosong")
  sorted_data = sorted(data)
  for i in range(len(sorted_data)):
    if sorted_data[i] == sorted_data[i + 1]:
      return sorted_data[i]

def variance(data: list[float]) -> float:
  if len(data) == 0:
    raise ValueError("List tidak boleh kosong")
  n = len(data)
  avg = sum(data)/len(data)
  sum_sq_diff = sum((x - avg) ** 2 for x in data)
  population_variance = sum_sq_diff / len(data) 
  return population_variance
  
def std_dev(data: list[float]) -> float:
  if len(data) == 0:
    raise ValueError("List tidak boleh kosong")
  return variance(data=data) ** 0.5

def data_range(data: list[float]) -> float:
  if len(data) == 0:
    raise ValueError("List tidak boleh kosong")
  return max(data) - min(data)
  
my_list = [4, 1, 2, 2, 3, 5]
print(f'Mean: {mean(my_list):.4f}')
print(f'Median: {median(my_list):.1f}')
print(f'Mode: {mode(my_list):.0f}')
print(f'Variance: {variance(my_list):.4f}')
print(f'Std Dev: {std_dev(my_list):.4f}')
print(f'Range: {data_range(my_list):.0f}')

```

### Soal 2: Normalisasi dan Standarisasi

```python
def min_max_normalize(data: list[float]) -> list[float]:
  if max(data) == min(data):
    raise ValueError("Semua nilai sama, tidak bisa dinormalisasi")
  new_list = [f"{x - min(data) / max(data) - min(data):.2f}" for x in data]
  return new_list

def z_score_standarize(data: list[float]) -> list[float]:
    mu = mean(data)
    sigma = std_dev(data)
    if sigma == 0:
        raise ValueError("Standar deviasi = 0, tidak bisa distandarisasi")
    return [(x - mu) / sigma for x in data]

def describe_scaling(data: list[float]) -> dict:
  my_dict = {
    "original_max": float(max(data)),
    "original_min": min(data),
    "original_mean": mean(data),
    "original_std": std_dev(data),
    "normalized_min": min(min_max_normalize(data)),
    "normalized_max": max(min_max_normalize(data)),
    "standardized_mean": mean(z_score_standarize(data=data)),
    "standardized_std": std_dev(z_score_standarize(data=data))
  }
  return my_dict

data_list = [1, 2, 3, 4, 5]
print(f'{min_max_normalize(data=data_list)}')
print(f'{z_score_standarize(data=data_list)}')
print(f'{describe_scaling(data=data_list)}')
```

### Soal 3: Manipulasi Matrix dengan List of Lists

```python
def matrix_add(A: list[list[float]], B: list[list[float]]) -> list[list[float]]:
  if len(A) != len(B):
    raise ValueError("Ukuran matrix tidak sama")
  result = [[A[i][j] + B[i][j] for j in range(len(A[0]))] for i in range(len(A))]
  return result

def matrix_multiply(A: list[list[float]], B: list[list[float]]) -> list[list[float]]:
  result = [[sum(a * b for a, b in zip(row_a, col_b)) for col_b in zip(*B)] for row_a in A]
  return result

def transpose(A: list[list[float]]) -> list[list[float]]:
  return [list(row) for row in zip(*A)]

def matrix_scalar_multiply(A: list[list[float]], scalar: float) -> list[list[float]]:
  result = [[element * scalar for element in row] for row in A]
  return result

print(matrix_add([[1, 2], [3, 4]], [[5, 6], [7, 8]]))
print(matrix_multiply([[1,2]], [[1],[2]]))
print(transpose([[1,2,3]]))
print(matrix_scalar_multiply([[1,2],[3,4]], 2))
```

### Soal 4: Fungsi Aktivasi Neural Network

```python
import math

def sigmoid(x: float) -> float:
  return 1 / (1 + math.exp(-x))

def relu(x: float) -> float:
  return max(0,x)

def tanh(x: float) -> float:
  return math.tanh(x)

def softmax(x: list[float]) -> list[float]:
  e_x = [math.exp(x) for x in x]
  sum_e_x = sum(e_x)
  return [x / sum_e_x for x in e_x]

def leaky_relu(x: float, alpha: float = 0.01) -> float:
  return x if x > 0 else alpha * x

def apply_activation(data: list[float], activation: str, **kwargs) -> list[float]:
  if activation == 'sigmoid':
    return [sigmoid(x) for x in data]
  elif activation == 'relu':
    return [relu(x) for x in data]
  elif activation == 'tanh':
    return [tanh(x) for x in data]
  elif activation == 'softmax':
    return softmax(data)
  elif activation == 'leaky_relu':
    alpha = kwargs.get("alpha", 0.01)
    return [x if x > 0 else alpha * x for x in data]
  else:
    raise ValueError(f"Aktivasi '{activation}' tidak dikenal")
  
print(apply_activation([-2, -1, 0, 1, 2], activation="relu"))
print(apply_activation([1,2,3], activation="softmax"))
print(apply_activation([-10,0,10], activation="leaky_relu"))
print(apply_activation([0], activation="tanh"))
# print(apply_activation([1,2], activation="unknown"))
```

### Soal 5: Encoding Fitur Kategorikal

```python
def label_encode(data: list[str]) -> tuple[list[int], dict]:
  unique_labels = sorted(list(set(data)))
  mapping = {label: i for i, label in enumerate(unique_labels)}
  return ([mapping[val] for val in data])

def one_hot_encode(data: list[str]) -> tuple[list[list[int]], list[str]]:
  unique_labels = sorted(list(set(data)))
  category_to_index = {cat: i for i, cat in enumerate(unique_labels)}
  encoded_matrix = []
  for item in data:
    vector = [0] * len(unique_labels)
    index = category_to_index[item]
    vector[index] = 1
    encoded_matrix.append(vector)    
  return (encoded_matrix, unique_labels)

def decode_label(encoded: list[int], mapping: dict) -> list[str]:
  max_index = encoded.index(max(encoded))
  return mapping[max_index]

data = ["cat", "dog", "bird", "cat", "dog"]
print(label_encode(data=data))
print(one_hot_encode(data=data))
print(decode_label(encoded=label_encode(data=data), mapping=data))
```

```python
first_number = int(input("Masukkan angka ke-1: "))
second_number = int(input("Masukkan angka ke-2: "))
third_number = int(input("Masukkan angka ke-3: "))

def factorial_iterative(n: int) -> int:
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

print(f'Faktorial dari {first_number} adalah: {factorial_iterative(first_number)}')
print(f'Faktorial dari {second_number} adalah: {factorial_iterative(second_number)}')
print(f'Faktorial dari {third_number} adalah: {factorial_iterative(third_number)}')
print(f'Total hitungan: 3 angka telah diproses')
```

### Soal 7: Dictionary untuk Data Karyawan

```python
data = []

def tambah_karyawan(nama: str, divisi: str, gaji: int):
  data_karyawan = {}
  data_karyawan['nama'] = nama
  data_karyawan['divisi'] = divisi
  data_karyawan['gaji'] = gaji
  data.append(data_karyawan)
  print(f'✓ {nama} (Divisi {divisi}, Gaji: {gaji}) ditambahkan')

def tampilkan_semua_karyawan(data):
  for i, karyawan in enumerate(data, 1):
    print(f'{i}. {karyawan["nama"]} - {karyawan["divisi"]} - Rp {karyawan["gaji"]}')

def cari_karyawan(nama, data):
  print(f'Dicari: {nama}')
  for item in data:
    if item['nama'] == nama:
      print(f'Ditemukan! {item['nama']} bekerja di {item['divisi']} dengan gaji Rp {item['gaji']}')
      break
    else:
      print(f'{nama} tidak ditemukan')

def update_gaji(nama: str, gaji_baru: int, data):
  print(f'Update gaji {nama} menjadi Rp {gaji_baru}')
  for item in data:
    if item['nama'] == nama:
      item['gaji'] = gaji_baru
      print(f'✓ Gaji {nama} berhasil diupdate')
      break
    else:
      print(f'Gaji {nama} gagal diupdate')

def hapus_karyawan(nama, data):
    print(f'Hapus {nama}')
    for item in data:
      if item['nama'] == nama:
        data.remove(item) 
        print(f'✓ {nama} berhasil dihapus dari sistem')
        return
    print(f'✗ {nama} gagal dihapus karena tidak ditemukan di sistem')
      
print('=== SISTEM MANAJEMEN KARYAWAN === \n')
print('Menambah data karyawan...')
tambah_karyawan("Budi", "IT", 8000000)
tambah_karyawan("Siti", "HR", 6500000)
tambah_karyawan("Ahmad", "Sales", 7000000)

print("\n=== SEMUA KARYAWAN ===")
tampilkan_semua_karyawan(data)

print(f'\n === CARI KARYAWAN ===')
cari_karyawan("Budi", data)

print('\n=== UPDATE GAJI ===')
update_gaji(nama='Budi', data=data, gaji_baru=9000000)

print('\n === HAPUS KARYAWAN ===')
hapus_karyawan(nama="Ahmad", data=data)
```

```python
def is_vowel(letter):
  return letter in ['a', 'e', 'i', 'o', 'u', 'y']

def score_words(words):
    score = 0
    for word in words:
        num_vowels = 0
        for letter in word:
            if is_vowel(letter):
                num_vowels += 1
        if num_vowels % 2 == 0:
            score += 2
        
    return score
  
print(score_words("programming is awesome"))
print(score_words("hacker book"))
```
