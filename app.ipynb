### Soal 6: Fungsi Faktorial dengan Validasi

```python
first_number = int(input("Masukkan angka ke-1: "))
second_number = int(input("Masukkan angka ke-2: "))
third_number = int(input("Masukkan angka ke-3: "))

def factorial_iterative(n: int) -> int:
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

print(f'Faktorial dari {first_number} adalah: {factorial_iterative(first_number)}')
print(f'Faktorial dari {second_number} adalah: {factorial_iterative(second_number)}')
print(f'Faktorial dari {third_number} adalah: {factorial_iterative(third_number)}')
print(f'Total hitungan: 3 angka telah diproses')
```

### Soal 7: Dictionary untuk Data Karyawan

```python
data = []

def tambah_karyawan(nama: str, divisi: str, gaji: int):
  data_karyawan = {}
  data_karyawan['nama'] = nama
  data_karyawan['divisi'] = divisi
  data_karyawan['gaji'] = gaji
  data.append(data_karyawan)
  print(f'✓ {nama} (Divisi {divisi}, Gaji: {gaji}) ditambahkan')

def tampilkan_semua_karyawan(data):
  for i, karyawan in enumerate(data, 1):
    print(f'{i}. {karyawan["nama"]} - {karyawan["divisi"]} - Rp {karyawan["gaji"]}')

def cari_karyawan(nama, data):
  print(f'Dicari: {nama}')
  for item in data:
    if item['nama'] == nama:
      print(f'Ditemukan! {item['nama']} bekerja di {item['divisi']} dengan gaji Rp {item['gaji']}')
      break
    else:
      print(f'{nama} tidak ditemukan')

def update_gaji(nama: str, gaji_baru: int, data):
  print(f'Update gaji {nama} menjadi Rp {gaji_baru}')
  for item in data:
    if item['nama'] == nama:
      item['gaji'] = gaji_baru
      print(f'✓ Gaji {nama} berhasil diupdate')
      break
    else:
      print(f'Gaji {nama} gagal diupdate')

def hapus_karyawan(nama, data):
    print(f'Hapus {nama}')
    for item in data:
      if item['nama'] == nama:
        data.remove(item) 
        print(f'✓ {nama} berhasil dihapus dari sistem')
        return
    print(f'✗ {nama} gagal dihapus karena tidak ditemukan di sistem')
      
print('=== SISTEM MANAJEMEN KARYAWAN === \n')
print('Menambah data karyawan...')
tambah_karyawan("Budi", "IT", 8000000)
tambah_karyawan("Siti", "HR", 6500000)
tambah_karyawan("Ahmad", "Sales", 7000000)

print("\n=== SEMUA KARYAWAN ===")
tampilkan_semua_karyawan(data)

print(f'\n === CARI KARYAWAN ===')
cari_karyawan("Budi", data)

print('\n=== UPDATE GAJI ===')
update_gaji(nama='Budi', data=data, gaji_baru=9000000)

print('\n === HAPUS KARYAWAN ===')
hapus_karyawan(nama="Ahmad", data=data)
```

### Soal 10: Try-Except untuk Input User

```python
first_number = int(input("Masukkan angka 1: "))
second_number = int(input("Masukkan angka 2: "))
operator = input("Pilih operasi (+, -, *, /): ")

try:
  if operator == '+':
    print(f'{first_number} {operator} {second_number} = {first_number + second_number}')
  elif operator == '-':
    print(f'{first_number} {operator} {second_number} = {first_number - second_number}')
  elif operator == '*':
    print(f'{first_number} {operator} {second_number} = {first_number * second_number}')
  elif operator == '/':
    print(f'{first_number} {operator} {second_number} = {first_number / second_number}')
except ValueError:
  raise ValueError("the input must be a number")
except ZeroDivisionError:
  raise ZeroDivisionError("cant divided by zero")
finally: 
  print(f'--- Operasi selesai ---')
```

#### Soal 11: Permainan Tebak Angka

```python
import random

random_number = random.randrange(1, 100, 1)

print(random_number)

print('=== PERMAINAN TEBAK ANGKA ===')
print('Saya sudah memikirkan sebuah angka antara 1-100')
print('Anda punya 10 kesempatan untuk menebaknya')

history = []

for i in range(10):
  guess_number = int(input(f'Tebakan ke-{i}:'))
  history.append(guess_number) # type: ignore
  try:
    if guess_number == random_number:
      print(f'BENAR! Angka yang saya pikir adalah {guess_number}')
      break
    else:
      if guess_number > random_number:
        print(f'Tebakan ke-{i+1}: {guess_number}')
        print(f'Terlalu besar!')
      else:
        print(f'Tebakan ke-{i+1}: {guess_number}')
        print(f'Terlalu kecil!')
  except ValueError:
    raise ValueError()

print(f'History Tebakan: {history}')
print(f'Jumlah Tebakan: {len(history)}') # type: ignore

score = 0

if len(history) < 3: # type: ignore
  score = 100
elif len(history) < 6: # type: ignore
  score = 75
elif len(history) < 10: # type: ignore
  score = 50
else:
  score = 0


print(f'Skor Anda: {score}')
```

```python
db_mahasiswa = {}

def tambah_mahasiswa(nim, nama, semester, gpa): # type: ignore
  if nim in db_mahasiswa:
    raise ValueError("cant adding the same NIM")
  else:
    db_mahasiswa[nim] = (nama, semester, gpa) # type: ignore
  print(f'✓ {nim} - {nama} (Semester {semester}, GPA {gpa}) ditambahkan')


def tampilkan_mahasiswa(nim, db):
  nama, semester, gpa = db[nim]
  return f'NIM: {nim} | Nama: {nama}  | Semester: {semester}  | GPA: {gpa}'

def update_gpa(nim, gpa_baru, db):
  nama, semester, gpa = db[nim]
  db[nim] = (nama, semester, gpa_baru)
  print(f'NIM {nim} - GPA diupdate dari {gpa} menjadi {gpa_baru}')
  print(f'✓ GPA {nama} berhasil diupdate')

def hapus_mahasiswa(nim, db):
  del db[nim]

def cari_mahasiswa_gpa_min(min_gpa, db):
  for nim, data in db.items():
    nama, semester, gpa = data
    if gpa >= min_gpa:
        print(f'{nim} - {nama}: {gpa}')

def statistik_gpa(db):
  gpas = [data[2] for data in db.values()]
  avg = sum(gpas) / len(gpas)
  max_gpa = max(gpas)
  min_gpa = min(gpas)
  print(f"Rata-rata GPA: {avg}")
  print(f"GPA Tertinggi: {max_gpa}")
  print(f"GPA Terendah: {min_gpa}")

print("=== DATABASE MAHASISWA === \n")

print("Menambah mahasiswa...")
tambah_mahasiswa("001", "Andi", 3, 3.75)
tambah_mahasiswa("002", "Bella", 2, 3.50)
tambah_mahasiswa("003", "Citra", 4, 3.90)
tambah_mahasiswa("004", "Doni", 1, 3.25)
tambah_mahasiswa("005", "Eka", 3, 3.80)

print("\n=== DATA MAHASISWA ===")
print(tampilkan_mahasiswa("001", db_mahasiswa))
print(tampilkan_mahasiswa("002", db_mahasiswa))
print(tampilkan_mahasiswa("003", db_mahasiswa))
print(tampilkan_mahasiswa("004", db_mahasiswa))
print(tampilkan_mahasiswa("005", db_mahasiswa))

print("\n=== STATISTIK GPA ===")
statistik_gpa(db_mahasiswa)

print("\n=== MAHASISWA DENGAN GPA >= 3.70 ===")
cari_mahasiswa_gpa_min(3.70, db_mahasiswa)

print("\n=== UPDATE GPA ===")
update_gpa("001", 3.80, db_mahasiswa)
```

```python
def is_vowel(letter):
  return letter in ['a', 'e', 'i', 'o', 'u', 'y']

def score_words(words):
    score = 0
    for word in words:
        num_vowels = 0
        for letter in word:
            if is_vowel(letter):
                num_vowels += 1
        if num_vowels % 2 == 0:
            score += 2
        
    return score
  
print(score_words("programming is awesome"))
print(score_words("hacker book"))
```
